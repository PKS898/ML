```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount ('/content/drive')
```

```python
# Загрузка изображения
image = Image.open('/content/drive/MyDrive/MII/Mash/kuro.jpg')

# Конвертация изображения в массив NumPy
input_array = np.array(image) / 255.0  # Нормализация значений пикселей до диапазона [0, 1]

# Расширение размерности массива для создания батча изображений
input_tensor = tf.expand_dims(input_array, axis=0)
```

# Определение операции свертки

```python
conv = Conv2D(filters=2, kernel_size=7, strides=5, padding='valid', use_bias=True)

# Применение операции свертки
output_tensor = conv(input_tensor)

# Получение массива NumPy из тензора
output_array = output_tensor.numpy()[0]
```

# Визуализация изображений

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(input_array)
plt.title('Input Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(output_array[:,:,0], cmap='gray')
plt.title('Output Channel 1')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(output_array[:,:,1], cmap='gray')
plt.title('Output Channel 2')
plt.axis('off')

plt.show()
```
